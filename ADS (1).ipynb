```python
#Illustrate a bar graph indicating the population in Canada by years.
import plotly.express as px
data_canada = px.data.gapminder().query("country == 'Canada'")
fig = px.bar(data_canada, x='year', y='pop')
fig.show()
```

```python
#Use Grouped bar graphs to show subgroups that need to be visualized on the
#graph. The subgroups should be differentiated by distinct colors.
#Ads:- https://colab.research.google.com/drive/16fdH8LSYLunPiG6MHiScLvrwnv7vosjl?usp=sharing 
#Link:- https://www.analyticsvidhya.com/blog/2021/12/12-data-plot-types-for-visualization/
#manual:- https://drive.google.com/file/d/1y65LOD-dFs5Ooo_GACAugOE1Jsmp-YUK/view?usp=sharing 
import seaborn as sb
df = sb.load_dataset('tips')
df = df.groupby(['size', 'sex']).agg(mean_total_bill=("total_bill", 'mean'))
df = df.reset_index()
sb.barplot(x="size", y="mean_total_bill", hue="sex", data=df)
```

```python
#Plot stacked bar graphs are used to show dataset subgroups.
import pandas
import matplotlib.pylab as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
df = pandas.DataFrame(dict(
   number=[2, 5, 1, 6, 3],
   count=[56, 21, 34, 36, 12],
   select=[29, 13, 17, 21, 8]
))
bar_plot1 = sns.barplot(x='number', y='count', data=df, label="count", color="red")
bar_plot2 = sns.barplot(x='number', y='select', data=df, label="select", color="green")
plt.legend(ncol=2, loc="upper right", frameon=True)
plt.show()
```

```python
#Use a line graph is used to visualize a trend in data over intervals of time.
import plotly.express as px
df = px.data.gapminder().query("country=='Canada'")
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in Canada')
fig.show()
```

```python
#Show the study of more than one variable over the same period using Multiple
#line graphs.
import plotly.express as px
df = px.data.gapminder().query("continent == 'Oceania'")
fig = px.line(df, x='year', y='lifeExp', color='country', symbol="country")
fig.show()
```

```python
#Illustrate numerical proportion, it is divided into slices. In a pie chart, for
#every slice, each of its arc lengths is proportional to the amount it represents.
import plotly.express as px
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
```

```python
#Draw a histogram to represent an approximation of the distribution of
#numerical data. By dividing the data into non-overlapping intervals.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from an external source
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
data = pd.read_csv(url)

# Create a histogram to show the distribution of total bill amounts
plt.hist(data['total_bill'], bins=20)
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bill Amounts')
plt.show()
```

```python
#Use Cartesian coordinates to display values for two variables for a set of
#data with the help of a scatter plot.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from an external source
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
data = pd.read_csv(url)

# Create a scatter plot to show the relationship between petal length and width
plt.scatter(data['petal_length'], data['petal_width'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Relationship between Petal Length and Width')
plt.show()
```

```python
#Plot a heatmap to show the correlation between features.
import pandas as pd
import seaborn as sns

# Load the dataset from an external source
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
data = pd.read_csv(url)

# Calculate the correlation matrix
corr = data.corr()

# Create a heatmap to visualize the correlation matrix
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#Find the mean, std deviation and other central tendencies
import pandas as pd

# Load the dataset from an external source
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
data = pd.read_csv(url)

# Calculate the mean of each numerical column
mean = data.mean()

# Calculate the standard deviation of each numerical column
std = data.std()

# Calculate the median of each numerical column
median = data.median()

# Calculate the mode of each numerical column
mode = data.mode()

# Print the results
print('Mean:')
print(mean)
print('\nStandard Deviation:')
print(std)
print('\nMedian:')
print(median)
print('\nMode:')
print(mode)
```
